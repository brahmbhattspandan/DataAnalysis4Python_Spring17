## Assignment 3_ Part 3_Question 1
- Use ‘cricket_matches’ data set.
- Calculate the average score for each team which host the game and win the game. 
- Remember that if a team hosts a game and wins the game, their score can be innings_1 runs or innings_2 runs. You have to check if  the host team won the game, check which innings they played in (innings_1 or innings_2), and take the runs scored in that innings. The final answer is the average score of each team satisfying the above condition.
- Display a few rows of the output **Use df.head()**
- Generate a csv output

```python
import pandas as pd
```

```python
df=pd.read_csv("Data/cricket_matches.csv")
```

```python
df.head()
```

```python
temp_df = df.loc[:,['home','winner','innings1','innings1_runs','innings2','innings2_runs']]
```

```python
temp_df.head()
```

```python
temp_df = temp_df.loc[temp_df['home']==temp_df['winner']]
temp_df.head()
```

```python
temp_df['score'] = temp_df[['innings1_runs','innings2_runs']].max(axis=1)
```

```python
temp_df.head()
```

```python
matches_df = round(temp_df.groupby(['home'],as_index=False)[['score']].mean(),2)
```

```python
matches_df = matches_df.rename(columns={'score':'average_score'})
final_df = matches_df.sort_values('average_score', ascending=False)
final_df.head()
```

```python
final_df.to_csv('outputs/cricket_matches_score.csv', index=False)
```
