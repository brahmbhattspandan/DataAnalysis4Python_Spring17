- Under package webtext, see what files are in it.
- Get content in file 'pirates.txt'.
- According to the number of lines for JACK SPARROW and WILL TURNER, find out who is the real leading man in this movie. 

```python
from nltk.corpus import webtext
```

```python
webtext.fileids()
```

```python
sentences = webtext.sents('pirates.txt')
len(sentences)
```

```python
pirates = webtext.words('pirates.txt') 
```

```python
print(pirates)
```

```python
jack_lines = 0
will_lines = 0
```

```python
for word in pirates:
    if word == 'SPARROW':
        jack_lines += 1
    elif word == 'WILL':
        will_lines += 1
```

```python
print('JACK SPARROW\'s lines are:', jack_lines)
print('WILL TURNER\'s lines are:', will_lines)
if jack_lines > will_lines:
    print('Pirates of the Carribean\'s leading actor is: Jack Sparrow')
else:
    print('Pirates of the Carribean\'s leading actor is: Will Turner')
```
