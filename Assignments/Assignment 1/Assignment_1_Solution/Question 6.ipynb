# Question 6
- Under Inaugural Address package
- For every president, find out how many unique words they used in speech. 
- We will just assume that President Bush means one person (technically it is incorrect but we will ignore it)

```python
from nltk.corpus import inaugural
```

```python
speech_list = inaugural.fileids()
```

```python
print(speech_list)
```

```python
president_words_list = []
```

```python
for file in speech_list:     # Merge the file first!
    name = file[5:-4]  
    words = inaugural.words(file)
#     all_words = list(x.lower() for x in inaugural.words(file))  # Get all word in lowercase
#     words = [lambda: x for x in all_words if x.isalpha() == True]  # Get all words only
    if len(president_words_list) == 0:
        president_words_list.append([name, words])
    else: 
        for president in president_words_list:
            if name == president[0]: 
                president[1] += (words)
                break
        else:                                              # Notice the else indentation
            president_words_list.append([name, words])
```

```python
print(len(president_words_list[0][1]))
```

```python
word_count_list = []
```

```python
for president in president_words_list:   # Then take unique words.
    name = president[0]
    words = set(president[1])
    word_count_list.append((name, len(words)))
```

```python
print('The Number of Unique Words Each President Used in Their Speeches:\n', word_count_list)
```

```python
import pandas as pd          # Create a simple DataFrame to show the results, will learn later.
```

```python
df = pd.DataFrame(word_count_list, index=range(1,35), columns=['Presidents', 'words count'])
```
